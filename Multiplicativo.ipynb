```python
import time
import matplotlib.pyplot as plt
import numpy as np
```

Pedir por consola el rango y la cantidad de números a generar

```python
range_start = int(input("Ingresa el inicio del rango: "))
range_end = int(input("Ingresa el fin del rango: "))
num_numbers = int(input("Ingresa la cantidad de números a generar: "))

print("Rango: ", range_start, " y ", range_end)
print("Cantidad: ", num_numbers)
```

Generador de números aleatorios usando el método congruencial multiplicativo

```python
def random_numbers(seed=None):
    # se hace una condicion para verificar si existe una semilla,
    # sino, se crea una obteniendo la hora del computador con la funcion .time()
    # si se cumple, se actualiza con el nuevo número generado, para que pueda ser
    # utilizada como semilla en la próxima iteración.
    if seed is None:
        seed = int(time.time())
        sem = seed
    
        
    a = 1103515245  # factor multiplicativo
    c = 12345 # incremento
    m = 2**31 - 1 # modulo

    semilla = sem
    succession = []
    numbers = []
    positions = []
    
    for i in range(1,num_numbers+1):
        xn = (a * seed + c) % m
        xu = xn / m
        
        # Generar numeros pseudoaleatorios
        numbers.append(int(range_start + (range_end - range_start) * xu))
        
        succession.append(xu)
        positions.append(i)
        seed = xn
    
    return semilla, numbers, succession, positions
```

```python
semilla, list_numbers, list_succession, list_positions = random_numbers()

print('Semilla inicial:', semilla)
print('Numeros generados: ', list_numbers)
print('Intervalos: ', list_succession)
print('posiciones: ', list_positions)
```

Función para buscar patrones en la lista de números generados

```python
# Se crea una funcion para iterar la lista de numeros generados
# y buscar un patron si es que existe dentro de la secuencia generada
def search_pattern(numbers):
    for i in range(1, len(numbers)):
        pattern = numbers[:i]
        repetitions = len(numbers) // len(pattern)
        if numbers == pattern * repetitions:
            return pattern
    return None
```

Verificar si existe un patron en la secuencia

```python
print("Numero generados: ", list_numbers)

pattern = search_pattern(list_numbers)
if pattern is not None:
    print("Se encontró el patrón en la secuencia:", pattern)
else:
    print("\033[91mNo se encontró ningún patrón en la secuencia\033[0m")
```

### Generacion de Diagramas

Diagrama de Dispersion

```python
from matplotlib import colors


x = list_positions[:-1]
y = list_succession[1:]
plt.scatter(x, y)
plt.title("Diagrama de dispersión")
plt.xlabel("Posiciones")
plt.ylabel("Sucesion")
plt.show()
```

Histograma

```python

fig, ax = plt.subplots()
n, bins, patches = ax.hist(list_succession, bins=10, density=True, color='cornflowerblue', edgecolor='black')

# Personalizar colores de los parches
fracs = n / n.max()
norm = plt.Normalize(fracs.min(), fracs.max())
for patch, frac in zip(patches, fracs):
    color = plt.cm.viridis(norm(frac))
    patch.set_facecolor(color)

# Mostrar el histograma
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

Diagrama de Autocorrelacion

```python
# Calcular autocorrelación
autocorr_data = np.correlate(list_succession, list_succession, mode="full")

# Calcular media de los numeros
media_data = np.mean(list_succession)

# Configuración
# Gráfico de autocorrelación
plt.stem(
    (list_succession - media_data),
    use_line_collection=True,
    linefmt="gray",
    markerfmt="bo",
    basefmt="r-",
)
plt.title("Autocorrelación")
plt.grid(True)
plt.show()
```
